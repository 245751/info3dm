```python
# 適当なデータを用意
import numpy as np
datasets = np.array([[4,7], [8,10], [13,11], [17,14]])
x = datasets[:,0]
y = datasets[:,1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color="black",label="dataset")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")
plt.show()
```

```python
def fanction(x):
    return (x-10)**2

x2=np.linspace(4,17,5)
y2=fanction(x2)
plt.plot(x2,y2,color="blue",linewidth=3,label="plot_example")
plt.legend(loc="best")
plt.show()
```

```python
x2=np.linspace(4,17,100)
y2=fanction(x2)
plt.plot(x2,y2,color="blue",linewidth=3,label="plot_example")
plt.legend(loc="best")
plt.show()
```

```python
#複数のグラフを1つに描画
plt.scatter(x, y, color="black", label="dataset")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x2, y2, color="blue", linewidth=3, label="plot example")
plt.legend(loc="best")
plt.show()
```
